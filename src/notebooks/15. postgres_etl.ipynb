```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("PySarkExp1").config("spark.jars", "./libs/postgresql-42.7.7.jar").getOrCreate()
```

```python
spark
```

```python
df = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://localhost:5432/commons") \
    .option("dbtable", "user_mgmt.users") \
    .option("user", "postgres") \
    .option("password", "sachin") \
    .option("driver", "org.postgresql.Driver") \
    .load()
```
